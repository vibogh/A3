```python
import folium

m = folium.Map([47.82328,13.03961], zoom_start=18, tiles='cartodbpositron')

folium.Marker(location=[47.82365,13.03934],popup='Department of Geoinformatics',tooltip='<strong>University of Salzburg</strong>',icon=folium.Icon(color='blue',icon='graduation-cap',prefix='fa')).add_to(m)

folium.GeoJson(
    "TechnoZ.json",
    name='TECHNOZ',
    style_function=lambda x: {'fillColor': 'green', 'color': 'blue', 'weight': '1'}
).add_to(m)


folium.raster_layers.ImageOverlay(
        name="TechnoZ",
        image='TechnoZ.png',
        bounds=[[47.823995, 13.038058], [47.82241575943387, 13.040649116504666]],
        opacity=0.5
    ).add_to(m)
m
```
